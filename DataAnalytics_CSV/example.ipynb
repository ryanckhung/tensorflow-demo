```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
file_path = './data_wb.csv'
```

```python
df = pd.read_csv(file_path).dropna()

# convert the exp notation to integer
df['device_time'] = df['device_time'].astype('int64')*1000 
df["delta"] = df["client_time"] - df["device_time"]

print(df)
```

```python
df.plot(y=["gsr_data"], figsize=(32,3))
```

```python
print("Min: {}".format(df["gsr_data"].min()))
print("Max: {}".format(df["gsr_data"].max()))
print("Mean: {}".format(df["gsr_data"].mean()))
```

```python
print ("Number of data value =0: {}".format(len(df[df["gsr_data"]==0])))
print ("Number of data value =0.5: {}".format(len(df[df["gsr_data"]<=0.5])))
print ("Number of data value <=1: {}".format(len(df[df["gsr_data"]<=1])))
print ("Number of data value <=2: {}".format(len(df[df["gsr_data"]<=2])))
print ("Number of data value <=3: {}".format(len(df[df["gsr_data"]<=3])))
print ("Number of data value <=4: {}".format(len(df[df["gsr_data"]<=4])))
print ("Number of data value <=5: {}".format(len(df[df["gsr_data"]<=5])))
print ("Total number of sample: {}".format(len(df)))
```

```python
print ("Total number of sample: {}".format(len(df)))
print ("Number of sample with unique timestamp: {}".format(len(df["client_time"].unique())))
print ("Same timestamp will repeat: {} times.".format(len(df) / len(df["client_time"].unique())))
```

```python
print ("Time difference between WB and backend.")
print("Number of samplles with time difference >= 1s: {}".format(len(df[df["delta"]>=1000])))
print("Number of samplles with time difference < 1s: {}".format(len(df[df["delta"]<1000])))

print("\nMax: {}".format(df["delta"].max()))
print("Min: {}".format(df["delta"].min()))
print("Mean: {}".format(df["delta"].mean()))
print("Std: {}".format(df["delta"].std()))
```
