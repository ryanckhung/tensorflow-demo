```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Example 1

```python
x = np.linspace(0,10,100)
y = np.sin(x)
```

```python
plt.figure(figsize=(20, 3))
plt.plot(x,y)
```

```python
dx = np.diff(x)
dy = np.diff(y)
slope = dy/dx

print(slope.shape)
print(dx.shape)
print(x.shape)

plt.figure(figsize=(20, 3))
plt.plot(x, np.insert(slope, 0, slope[0]))
```

```python
plt.figure(figsize=(20, 3))
plt.plot(x, y, label="sin")
plt.plot(x, np.insert(slope, 0, slope[0]), label="slope of sin")
plt.legend()
plt.grid()
```

### Example 2

```python
x = np.arange(0,100)
y = 10*x + 2
```

```python
plt.plot(x,y)
```

```python
dx = np.diff(x)
dy = np.diff(y)
slope = dy/dx
slope = np.insert(slope, 0, slope[0])

plt.plot(x, slope)
```

### Example 3

```python
x = np.arange(0,100)
y = 3*x**3 + x**2 - 2*x + 3

plt.plot(x,y)
```

```python
dx = np.diff(x)
dy = np.diff(y)
slope = dy/dx

plt.figure(figsize=(20, 3))
slope = np.insert(slope, 0, slope[0])
plt.plot(x, slope)
plt.grid()
```
