```python
# Python example - Fourier transform using numpy.fft method
import numpy as np
import matplotlib.pyplot as plt

samplingFrequency  = 1/50

# Time points
timeline = np.arange(0, 10, samplingFrequency);

# frequence of the sin wave
f1 = 5
f2 = 3

# Create two sine waves
amplitude1 = np.sin(2*np.pi*f1*timeline) * 2 
amplitude2 = np.sin(2*np.pi*f2*timeline) * 1
```

```python
plt.figure(figsize=(20,5))
plt.title("Sine Wave with frequence {}Hz".format(f1))
plt.xlabel("seconds")
plt.ylabel("Amplitude")
plt.plot(timeline, amplitude1)
```

```python
plt.figure(figsize=(20,5))
plt.title("Sine Wave with frequence {}Hz".format(f1))
plt.xlabel("seconds")
plt.ylabel("Amplitude")
plt.plot(timeline, amplitude2)
```

```python
# Add the sine waves
amplitude = amplitude1 + amplitude2 

plt.figure(figsize=(20,5))
plt.title("Sine Wave with frequence {}Hz".format(f1))
plt.xlabel("seconds")
plt.ylabel("Amplitude")
plt.plot(timeline, amplitude)
```

```python
# Frequency domain representation

def fft(amplitude, samplingFrequency):
    fourierTransform = abs(np.fft.fft(amplitude))        
    fourierTransform = fourierTransform[range(int(len(amplitude)/2))] # Exclude sampling frequency

    tpCount     = len(amplitude)
    values      = np.arange(int(tpCount/2))
    timePeriod  = tpCount*samplingFrequency
    frequencies = (values/timePeriod)
    return (frequencies, fourierTransform)


fft = fft(amplitude, samplingFrequency)

plt.figure(figsize=(20,5))
plt.title("Fourier transform depicting the frequency components")
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.plot(fft[0], fft[1])
```
